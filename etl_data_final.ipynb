### Data extraction & Cleanup - US Pollution dataset

```python
# Read the file
file= "Resources/pollution_us_2000_2016.csv"
df = pd.read_csv(file, encoding="ISO-8859-1")
df.head(2)
```

```python
# Clean unwanted columns
df2=df.drop(['Unnamed: 0','State Code','County Code','Site Num', 'Address', 'County',
        'O3 Units', 'O3 Mean', 'O3 1st Max Value','O3 1st Max Hour', 'O3 AQI',
        'NO2 1st Max Value','NO2 1st Max Hour','NO2 AQI',
        'SO2 1st Max Value', 'SO2 1st Max Hour', 'SO2 AQI', 
        'CO 1st Max Value', 'CO 1st Max Hour', 'CO AQI','NO2 Units', 'SO2 Units', 'CO Units' ], axis = 1) 

df2.head(5)
```

```python
# Split the year column and delete part of (%m-%d)
split_year_df = df2["Date Local"].str.split("-", n = 1, expand = True)
df2["Year"]= split_year_df[0]
df2["month-day"]= split_year_df[1]
del df2["month-day"]
df2.head()
```

```python
df3=df2.drop(["Date Local"], axis = 1) 
df3.head()
```

```python
#Convert data from parts per billion to grams/ton
NO2_mean = [(x* 0.001) for x in df3["NO2 Mean"]]
SO2_mean = [(x* 0.001) for x in df3["SO2 Mean"]]
CO_mean = [(x* 0.001) for x in df3["CO Mean"]]
```

```python
df3["no2_mean"] = NO2_mean
df3["so2_mean"] = SO2_mean
df3["co_mean"] = CO_mean
df3.head()
```

```python
df4=df3.drop(['NO2 Mean','SO2 Mean', 'CO Mean' ], axis = 1) 
```

```python
renamed_df= df4.rename(columns={"State": "state", "City": "city", "Year": "years"})                             
renamed_df.head() 
```

```python
# Create a new df
ordered_df=renamed_df[["years", "state", "city", "no2_mean", "so2_mean", "co_mean"]]
ordered_df.head()
```

```python
# Group dataframe by city, state and years
grouped_df_1 = ordered_df.groupby(["city","state", "years"])
# Create a new dataframe to aggregate the Mean columns
grouped_df_1_mean=pd.DataFrame(grouped_df_1["no2_mean", "so2_mean", "co_mean"].mean().reset_index())
grouped_df_1_mean.head()
```

```python
# Group dataframe by years, state and city
grouped_df_2 = ordered_df.groupby([ "years","state", "city"])
# Create a new dataframe to aggregate the Mean columns
grouped_df_2_mean=pd.DataFrame(grouped_df_2["no2_mean", "so2_mean", "co_mean"].mean().reset_index())
grouped_df_2_mean.head()
```

```python
# Create a new dataframe to aggregate the Mean columns.
year_df = ordered_df.groupby(["years"])
mean_df=pd.DataFrame(year_df["no2_mean", "so2_mean", "co_mean"].mean().reset_index())
#print(mean_df)
mean_df.head(10)
```

```python
year_state_df = ordered_df.groupby(['years','state']).agg({'no2_mean' : 'mean', 'so2_mean' : 'mean', 'co_mean': 'mean'})
print(year_state_df.head(100))
```

```python
year_city_df = ordered_df.groupby(['years','city']).agg({'no2_mean' : 'mean', 'so2_mean' : 'mean', 'co_mean': 'mean'})
print(year_city_df.head(100))
```

```python
#Generate new csv files from the dataframes
mean_df.to_csv("Resources/mod_mean_us_pollution.csv",index = False)
year_state_df.to_csv("Resources/mod_pollution_year_state.csv")
year_city_df.to_csv("Resources/mod_pollution_year_city.csv")
```

### Data extraction & Cleanup - OECD Air Emissions by source dataset

```python
#set path and read csv into a DataFrame
path = "Resources/air_emissons_by_source.csv"
df=pd.read_csv(path)
df.head()
```

```python
#print column names
for col in df.columns:
    print(col)
```

```python
#parse down df to only keep the columns we need 
df=df[["COU", "POL", "YEA", "Unit", "Value", "Variable", "VAR"]]
df.head()
```

```python
#remove rows with unnecessary data
pollutants_extra=df[(df['POL']=="NMVOC") | (df['POL']=="PM10") | (df['POL']=="PM2-5")].index
df=df.drop(pollutants_extra)
df.head()
```

```python
#rename values so they will match the other dataset
df['POL']= df['POL'].replace({'SOX': 'SO2', 'NOX': 'NO2'})
df.head(25)
```

```python
#drop any rows with null data
clean_df = df.dropna(axis = 0, how ='any')
clean_df.head(25)
```

```python
#parse down df
clean_df=clean_df[["POL", "Value", "Variable", "VAR", "YEA"]]
clean_df.head()
```

```python
#rename columns
clean_df=clean_df.rename(columns={"YEA": "years", "VAR": "pol_var", "Variable": "pol_variable", "POL": "pol", "Value": "value"})
clean_df.head()
```

```python
#so2 df
so2_df=clean_df[(clean_df["pol"]=="SO2")]
so2_df.head(17)
```

```python
#co df
co_df=clean_df[(clean_df["pol"]=="CO")]
co_df.head(17)
```

```python
#no2 df
no2_df=clean_df[(clean_df["pol"]=="NO2")]
no2_df.head(17)
```

```python
#write to individual csv files
so2_df.to_csv("Resources/mod_cleaned_air_emission_by_source_so2.csv", index=False)
co_df.to_csv("Resources/mod_cleaned_air_emission_by_source_co.csv", index=False)
no2_df.to_csv("Resources/mod_cleaned_air_emission_by_source_no2.csv", index=False)
```
