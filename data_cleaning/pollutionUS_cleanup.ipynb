```python
# Dependencies
import pandas as pd
```

```python
# Read the file
file= "Resources/pollution_us_2000_2016.csv"
df = pd.read_csv(file, encoding="ISO-8859-1")
df.head(2)
```

```python
# Check the columns names
list(df.columns)
```

```python
# Check null rows
df.count()
```

```python
# Drop duplicate rows
dropped_df=df.drop_duplicates(keep="last")
print(dropped_df)
```

```python
#Check the rows
unique = df["Date Local"].unique()
print(unique)
#Note: All rows are unique
```

```python
# Clean unwanted columns
df2=df.drop(['Unnamed: 0','State Code','County Code','Site Num', 'Address', 'County',
        'O3 Units', 'O3 Mean', 'O3 1st Max Value','O3 1st Max Hour', 'O3 AQI',
        'NO2 1st Max Value','NO2 1st Max Hour','NO2 AQI',
        'SO2 1st Max Value', 'SO2 1st Max Hour', 'SO2 AQI', 
        'CO 1st Max Value', 'CO 1st Max Hour', 'CO AQI','NO2 Units', 'SO2 Units', 'CO Units' ], axis = 1) 

df2.head(5)
```

```python
# Split the year column and delete part of (%m-%d)
split_year_df = df2["Date Local"].str.split("-", n = 1, expand = True)
df2["Year"]= split_year_df[0]
df2["month-day"]= split_year_df[1]
del df2["month-day"]
df2.head()
```

```python
# Drop "Date Local" column
df3=df2.drop(["Date Local"], axis = 1) 
df3.head()
```

```python
#Convert data from ppb to g/ton
NO2_mean = [(x* 0.001) for x in df3["NO2 Mean"]]
SO2_mean = [(x* 0.001) for x in df3["SO2 Mean"]]
CO_mean = [(x* 0.001) for x in df3["CO Mean"]]
```

```python
# Renamed with lower case
df3["no2_mean"] = NO2_mean
df3["so2_mean"] = SO2_mean
df3["co_mean"] = CO_mean
df3.head()
```

```python
#Drop unwanted columns
df4=df3.drop(['NO2 Mean','SO2 Mean', 'CO Mean' ], axis = 1) 
```

```python
# Rename the colums
renamed_df= df4.rename(columns={"State": "state", "City": "city", "Year": "years"})                                  
renamed_df.head()                        
```

```python
# Order the renamed_df
ordered_df=renamed_df[["years", "state", "city", "no2_mean", "so2_mean", "co_mean"]]
ordered_df.head()
```

```python
# Group dataframe by city, state and years and 
grouped_df_1 = ordered_df.groupby(["city","state", "years"])
# Create a new dataframe to aggregate the Mean columns.
grouped_df_1_mean=pd.DataFrame(grouped_df_1["no2_mean", "so2_mean", "co_mean"].mean().reset_index())
grouped_df_1_mean.head()
```

```python
# Group dataframe by city, state and years
grouped_df_2 = ordered_df.groupby([ "years","state", "city"])
# # Create a new dataframe to aggregate the Mean columns.
grouped_df_2_mean=pd.DataFrame(grouped_df_2["no2_mean", "so2_mean", "co_mean"].mean().reset_index())
grouped_df_2_mean.head()
```

```python
# Create a new dataframe to aggregate the Mean columns.
year_df = ordered_df.groupby(["years"])
mean_df=pd.DataFrame(year_df["no2_mean", "so2_mean", "co_mean"].mean().reset_index())
#print(mean_df)
mean_df.head(10)
```
